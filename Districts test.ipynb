```python
#read in POI
aoi = gpd.read_file(r'C:\Users\ktm0001\OneDrive - Dallas College\Documents\GIS training\Introduction to GeoPandas\Shapefiles\area_of_interest.shp')
aoi.plot(color = 'orange')
```

```python
#plot figures side by side
import matplotlib.pyplot as plt
fig, (ax1, ax2) = plt.subplots(ncols = 2)
D.plot(cmap = 'hsv', edgecolor = 'black', 
       column = 'district', 
       ax = ax1)
aoi.plot(color = 'orange', 
         ax = ax2)
```

```python
#layer maps 
#axes are layered on top of one another
fig, ax = plt.subplots()
D.plot(cmap = 'hsv', edgecolor = 'black', c
       olumn = 'district', 
       ax = ax)
aoi.plot(color = 'none', 
         ax = ax, 
         edgecolor = 'black')
```

```python
#read in atm locations
atm = gpd.read_file(r'C:\Users\ktm0001\OneDrive - Dallas College\Documents\GIS training\Introduction to GeoPandas\Shapefiles\atms.shp')
#layer all three areas 
fig, ax = plt.subplots()
D.plot(cmap = 'hsv', edgecolor = 'black',
       column = 'district', 
       ax = ax)
aoi.plot(color = 'none', 
         ax = ax, 
         edgecolor = 'black')
atm.plot(ax = ax,
         color = 'black',
         markersize = 12)
```
